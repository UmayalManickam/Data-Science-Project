```python
#IMPORTING LIBRARIES
import numpy as np
import tensorflow as tf
import neurolab as nl
import matplotlib.pyplot as plt
```

### TASK 1

#### Activation Function

```python
def tanh(x):
   return np.tanh(x);
def sig(x):
  return 1/(1+np.exp(-x));
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x), axis=0);
def custom(x):
  return  3* np.square(x) + 5;
```

A)Tanh-It returns the value (1-exp(-2x))/(1+exp(-2x)) and the value returned will be lies in between -1 to 1.

B)Sigmoid-It returns 1/(1+exp(-x)),where the values lies between 0 and 1.

C)Softmax-Compute softmax values for each sets of scores in x. Softmax turns logits, the numeric output of the last linear layer of a multi-class classification neural network into probabilities.

### TASK 2

```python
#A)GENERATING 130 TRAINING DATA
x = np.linspace(-15, 15,130)
fig,ax=plt.subplots(1,4,figsize=(20,5))

ax[0].plot(x, tanh(x))
ax[0].set_title('Activation Function :Tanh')

ax[1].plot(x, sig(x))
ax[1].set_title('Activation Function :Sigmoid')

ax[2].plot(x, softmax(x))
ax[2].set_title('Activation Function :Softmax')

ax[3].plot(x, custom(x))
ax[3].set_title('Activation Function :Custom')

fig.tight_layout()
plt.show()
```

### CHOOSING ONE ACTIVATION FUNCTION FROM THE ABOVE FOUR

```python
#B)GENERATING LABEL DATA
labels=tanh(x)
print(labels.shape)
```

```python
#C)PLOTTING THE INPUT DATA
plt.scatter(x,labels)
plt.show()
```

```python
x=x.reshape(130,1)
labels=labels.reshape(130,1)
# D)MULTILAYER NEURAL NETWORK
perceptron = nl.net.newff([[-15,15]], [10,6, 1])
# Set the training algorithm to gradient descent
perceptron.trainf = nl.train.train_gd
# Train the perceptron using the x with labels of tanh activation function
error_progress = perceptron.train(x, labels, epochs=1000, show=100)
# Generating output by running on training data points
output=perceptron.sim(x) 
```

```python
labels1=sig(x)
# Train the perceptron using the x with labels of Sigmoid activation function
error_progress1 = perceptron.train(x, labels1, epochs=1000, show=100)
# Generating output by running on training data points
output1=perceptron.sim(x)
```

```python
labels2=softmax(x)
# Train the perceptron using the x with labels of softmax activation function
error_progress2 = perceptron.train(x, labels2, epochs=1000, show=100)
# Generating output by running on training data points
output2=perceptron.sim(x) 
```

```python
labels3=custom(x)
# Train the perceptron using the x with labels of softmax activation function
error_progress3 = perceptron.train(x, labels3, epochs=1000, show=100)
# Generating output by running on training data points
output3=perceptron.sim(x) 
```

```python
# Plot training errors
fig,ax=plt.subplots(1,4,figsize=(20,5))
ax[0].plot(error_progress1)
ax[0].set_title('Training error progress-tanh')

ax[1].plot(error_progress2)
ax[1].set_title('Training error progress-Softmax')

ax[2].plot(error_progress3)
ax[2].set_title('Training error progress-Custom')

ax[3].plot(error_progress)
ax[3].set_title('Training error progress-Sigmoid')

plt.show()
```

```python
# Plot output
fig,ax=plt.subplots(1,4,figsize=(20,5))
ax[0].plot(output1)
ax[0].set_title('Output-tanh')

ax[1].plot(output2)
ax[1].set_title('Output-Softmax')

ax[2].plot(output3)
ax[2].set_title('Output-Custom')

ax[3].plot(output)
ax[3].set_title('Output-Sigmoid')

plt.show()
```

```python
#PLOT ACTUAL VS PREDICTED
plt.plot(x, labels, '-', x, tanh(x), '.', x, output, 'p')
plt.title('Actual vs predicted')
plt.show()
```
