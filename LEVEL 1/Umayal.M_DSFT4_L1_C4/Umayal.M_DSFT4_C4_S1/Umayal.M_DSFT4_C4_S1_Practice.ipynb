```python
# TASK 1
import numpy as np
HR=np.arange(1,501)
Finance=np.arange(501,1001)
IT=np.arange(1001,1501)
Sales=np.arange(1501,2001)
print(HR)
print(Finance)
print(IT)
print(Sales)
```

```python
# TASK 2
HR_salary=np.random.randint(25000,50001,500)
print('HR salary=',HR_salary)
print()
Finance_salary=HR_salary*1.25
print('Finance salary=',Finance_salary)
print()
Sales_salary=Finance_salary*0.5
print('Sales salary=',Sales_salary)
print()
IT_salary=HR_salary+5000
print('IT salary=',IT_salary)
```

```python
# TASK 3
a_sum=np.sum(HR_salary)
print('Sum of HR=',a_sum)
b_sum=np.sum(Finance_salary)
print('Sum of Finance=',b_sum)
c_sum=np.sum(Sales_salary)
print('Sum of Sales=',c_sum)
d_sum=np.sum(IT_salary)
print('Sum of IT=',d_sum)
if a_sum>b_sum:
    print('HR is costing higher')
elif b_sum>c_sum:
    print('Finance is costing higher')
elif c_sum>d_sum:
    print('Sales is costing higher')
elif d_sum>a_sum:
    print('IT is costing higher')
    
```

```python
# TASK 4
HR_high=np.sort(HR_salary)[::-1]
a=HR_high[0:10]
print('Highest salary of HR department=',a)
Finance_high=np.sort(Finance_salary)[::-1]
b=Finance_high[0:10]
print('Highest salary of Finance department=',b)
Sales_high=np.sort(Sales_salary)[::-1]
c=Sales_high[0:10]
print('Highest salary of Sales department=',c)
IT_high=np.sort(IT_salary)[::-1]
d=IT_high[0:10]
print('Highest salary of IT department=',d)
print()
a=np.sum(a)
print('Sum of HR=',a)
b=np.sum(b)
print('Sum of Finance=',b)
c=np.sum(c)
print('Sum of Sales=',c)
d=np.sum(d)
print('Sum of IT=',d)
```

```python
# TASK 5
a=np.count_nonzero(HR_salary>45000)
print('No.of employee in HR whose salary is higher than 45,000 is',a)
b=np.count_nonzero(Finance_salary>45000)
print('No.of employee in Finance whose salary is higher than 45,000 is',b)
c=np.count_nonzero(Sales_salary>45000)
print('No.of employee in Sales whose salary is higher than 45,000 is',c)
d=np.count_nonzero(IT_salary>45000)
print('No.of employee in IT whose salary is higher than 45,000 is',d)
print()
print('Total no.of employee whose salary is higher than 45,000 is',a+b+c+d)
```
