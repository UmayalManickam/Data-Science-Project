```python
# TASK 2
#BMW selling price in 2010 and 2015(diff)
  #condition>4.5, avg selling price in terms of percent(diff)
```

```python
# TASK 5
   #total sales of cars with SUV bodies on 'Fri Dec 19,2014, 09:00:00 GMT-0800' and 'Thu Jan 15,2015, 04:30:00 GMT-0800 (PST)'
   #sales of car(%) in this period
```

```python
#filter
f1=df[(df.body=='SUV')&(df.saledate=='Thu Jan 15 2015 04:30:00 GMT-0800 (PST)')]
#f1
```

```python
f2=df[(df.body=='SUV')&(df.saledate=='Fri Dec 19 2014 09:00:00 GMT-0800 (PST)')]
#f2
```

```python
#total sales
sale1=f1['sellingprice'].sum()
print('Total sale of cars with SUV bodies during Thu Jan 15 2015=',sale1)
sale2=f2['sellingprice'].sum()
print('Total sale of cars with SUV bodies during Fri Dec 19 2014=',sale2)
```

```python
#percentage
dif=((sale2-sale1)/sale2)*100
print('The total sales of car with SUV body in Thu Jan 15 2015 04:30:00 GMT-0800 (PST) has decreased by',dif,'%',' while comparing with Fri Dec 19 2014 09:00:00 GMT-0800 (PST)')
```
