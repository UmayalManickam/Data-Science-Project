```python
# TASK 1           
from numpy import genfromtxt
main=genfromtxt("b.csv",delimiter=',',skip_header=1)
print(main)
X1=main.T
Y1=np.count_nonzero(X1[1]>320)
print('Total no.of students whose GRE score is greater than 320 is',Y1)
```

```python
# TASK 2

Z2=np.where(main[:,2]>110)
print('Admission possibility of students whose TOEFL score is more than 110 is')
print(main[Z2,0:9:8])
```

```python
# TASK 3
x3= np.array(sorted(main, key = lambda x: x[1], reverse = True))
print("Top 11 GRE toppers:")
print(x3[0:11,[0,1]])
y3= np.array(sorted(main, key = lambda x: x[2], reverse = True))
print("Top 8 TOEFL toppers:")
print(y3[0:9,[0,2]])
```

```python
# TASK 4
x4=np.where((main[:,1]>=300) & (main[:,1]<=305) &(main[:,4]==3.5) & (main[:,7]==1))
print('According to the given condition displaying Serial no and TOEFL scores:')
print(main[:,[0,2]][x4])
                               
```

```python
# TASK 6
a=np.where(main[:,9]>9000)
print('SOP and LOR score for students who are getting scholarship more than 9000 are:')
print(main[:,[0,4,5]][a])
```

```python
# TASK 7
x7=np.where((main[:,1]>315) & (main[:,2]>115)& (main[:,7]==1))
print('Serial number of students who are eligible for $1000 scholarship are:')
print(main[:,0][x7])
print('count=',np.size(x7))
```

```python
#updating
main[:,9][x7]=1000
main[:,[0,9]][x7]
```

```python
# TASK 8
cgpa=np.where((main[:,6]>=8.0) & (main[:,6]<=8.5))
print('Total no.of students who have obtained CGPA from 8.0 to 8.5 are',np.size(cgpa))
```

```python
# TASK 9
initial=main[0:50]
avg_sop=np.mean(initial[:,4])
avg_lor=np.mean(initial[:,5])
print('Average of SOP score is more than LOR score by',((avg_sop-avg_lor)/avg_sop)*100,'%')
```

```python
# TASK 10
#spread/distribution of scholarship amount
s1=np.mean(main[:,9])
print('Average value of scholarship amount=',s1)
s2=np.max(main[:,9])
print('Maximum value of scholarship amount=',s2)
s3=np.min(main[:,9])

print('Minimum value of scholarship amount=',s3)
s4=np.median(main[:,9])
print('Median value of scholarship amount=',s4)
```

```python
# TASK 11
adm=np.where(main[:,8]>=0.95)
# np.size(adm)
print('GRE,TOEFL,CGPA required by students for atleast 95% admission possibility are:\n',main[:,[0,1,2,6]][adm],'\n')
print('Minimum GRE score required:',np.min(main[:,[1]][adm]),'\n')
print('Average GRE score required:',np.mean(main[:,[1]][adm]),'\n')
print('Minimum TOEFL score required:',np.min(main[:,[2]][adm]),'\n')
print('Average TOEFL score required:',np.mean(main[:,[2]][adm]),'\n')
print('Minimum CGPA required:',np.min(main[:,[6]][adm]),'\n')
print('Average CGPA required:',np.mean(main[:,[6]][adm]))
```
