```python
import pandas as pd
```

```python
#CREATING DATAFRAME FOR TABLE LOCATIONS
df_a=pd.read_sql_query('select * from hr.locations',db)
print(df_a)
```

```python
#CREATING DATAFRAME FOR TABLE EMPLOYEES
df_b=pd.read_sql_query('select * from hr.employees',db)
print(df_b)
```

```python
#CREATING DATAFRAME FOR TABLE DEPARTMENTS
df_c=pd.read_sql_query('select * from hr.departments',db)
print(df_c)
```

```python
#CREATING DATAFRAME FOR TABLE countries
df_d=pd.read_sql_query('select * from hr.countries',db)
print(df_d)
```

```python
#MERGING EMPLOYEES AND DEPARTMENTS
m1=pd.merge(df_b,df_c,on='department_id')
#m1
```

```python
#MERGING LOCATIONS
m2=pd.merge(df_a,m1,on='location_id')
#m2
```

```python
#MERGING COUNTRIES
m3=pd.merge(df_d,m2,on='country_id')
m3
```

```python
# TASK 3
import matplotlib.pyplot as plt
```

```python
#CITY AND EMPLOYEES
u=m3.groupby(by='city')['employee_id'].count()
#u
u.plot(x='city',kind='bar')
plt.title('City wise Employee count')
plt.xlabel('City')
plt.ylabel('No.of Employee')
plt.show()
```
