```python
while True:
    print("       ESSENTIAL ITEMS AVAILABLE")
    print()
    print('S.NO  ITEM        QUANTITY    COST/ITEM')
    print()
    print('1     Biscuits    5           20.50')
    print('2     Cereals     10          90')
    print('3     Chicken     20          200')
    print('4     Oats        20          39.99')
    print('5     Rice        30          79.99')
    print()
    n=int(input('Enter no.of items: '));
    print()
    Total_cost=0
    items = []
    for i in range(0,n):
        num=int(input('Enter the S.NO of the item required: '))
        print()
        if num==1:
            num1=int(input('Biscuits required: '))
            if num1>5:
                print('OUT OF STOCK')
            else:
                c1=20.50*num1
                Total_cost+=c1
                items.append((num,num1))
        elif num==2:
            num2=int(input('Cereals required: '))
            if num2>10:
                print('OUT OF STOCK')
            else:
                c2=90*num2
                Total_cost+=c2
                items.append((num,num2))
        elif num==3:
            num3=int(input('Chicken required: '))
            if num3>20:
                print('OUT OF STOCK')
            else:
                c3=200*num3
                Total_cost+=c3
                items.append((num,num3))
        elif num==4:
            num4=int(input('Oats required: '))
            if num4>20:
                print('OUT OF STOCK')
            else:
                c4=39.99*num4
                Total_cost+=c4
                items.append((num,num4))
        elif num==5:
            num5=int(input('Rice required: '))
            if num5>30:
                print('OUT OF STOCK')
            else:
                c5=79.99*num5
                Total_cost+=c5
                items.append((num,num5))
                
        else:
            print('Enter number between 1 to 5')
        print()
    #print(items)
############################################################################
    print("FILL IN YOUR DELIVERY DETAILS")
    print()
    name=input('Enter your name: ')
    address=input('Enter your address: ')
    dis=int(input('Enter the distance from the store: '))
    print()
    d1=10
    d2=25
    d3=35
    d=0
    if dis<=5:
        print('Your delivery charge is: ',d1)
        d=d1
    elif dis>5 and dis<=10:
        print('Your delivery charge is: ',d2)
        d=d2
    elif dis>10 and dis<20:
        print('Your delivery charge is: ',d3)
        d=d3
    else:
        print('Sorry! Your destination is too away')
#############################################################################
    print()
    print()
    print()
    print("        BILL      ")
    print()
    print('S.NO  ITEM        QUANTITY    TOTAL_COST')
    print()
    for s,q in items:
        if s==1:
            print(s,'   ','Biscuits    ',q,'         ',c1)
        elif s==2:
            print(s,'   ','Cereals     ',q,'         ',c2)
        elif s==3:
            print(s,'   ','Chicken     ',q,'         ',c3)
        elif s==4:
            print(s,'   ','Oats        ',q,'         ',c4)
        elif s==5:
            print(s,'   ','Rice        ',q,'         ',c5)
    print()
    final=(Total_cost+d)
    print("Total Item Cost: ",Total_cost)
    print()
    print("Total Bill Amount: ",final)
    print()
    print('Name: ',name)
    print('Address: ',address)
    print()
    if input('Do you want to continue shopping?')!='y':
        print('HAVE A NICE DAY!!')
        break
    print()
```
