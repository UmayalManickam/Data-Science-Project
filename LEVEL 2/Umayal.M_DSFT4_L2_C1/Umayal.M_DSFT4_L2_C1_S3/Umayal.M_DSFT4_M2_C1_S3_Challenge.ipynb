```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

### TASK 1

```python
url='https://store.steampowered.com/tags/en/Action/#p=0&tab=NewReleases'
response=requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content)
soup.prettify
```

```python
a=soup.find_all('div',class_="tab_item_name")
#a[0:10]
```

```python
games_title= [title.text for title in a]
#games_title
```

```python
b=soup.find_all('div',class_="discount_final_price")
#b[0:10]
games_price=[price.text for price in b]
#games_price
```

### TASK 2

```python
add=[games_title,games_price]
df=pd.DataFrame(add).T
df.head(10)
```

### TASK 4

```python
url1='https://store.steampowered.com/search/?filter=topsellers'
response1=requests.get(url1)
soup1=BeautifulSoup(response1.content)
#soup1
```

```python
aa=soup1.find_all('span',class_="title")
bb= [i.text for i in aa]
#bb
```

```python
c=pd.Series(bb)
c.value_counts()
```

```python
print('STAR WARS™: The Old Republic™ - Cartel Coins - is most widely sold')
```

### TASK 5

```python
d1=pd.DataFrame(bb,columns=['GAME TITLE'])
#d1
```

```python
cc=soup1.find_all('div',class_="col search_released responsive_secondrow")
dd= [i.text for i in cc]
d2=pd.DataFrame(dd,columns=['RELEASE DATE'])
#d2
```

```python
c=soup1.find_all('div',class_="col search_price responsive_secondrow")
d= [i.text.replace('\r','') for i in c]
d3=pd.DataFrame(d,columns=['GAME PRICE'])
d3['GAME PRICE']=d3['GAME PRICE'].str.replace('\n','')
#d3
```

```python
e=soup1.find_all('div',class_="col search_discount responsive_secondrow")
f= [i.text.replace('\n','') for i in e]
d4=pd.DataFrame(f,columns=['DISCOUNT'])
#d4
```

```python
DF=pd.concat([d1,d2,d4,d3], axis=1)
#DF
```

### TASK 6

```python
#EXTRACTING IMG URL IDs

image_urls=[]

url2 = 'https://store.steampowered.com/tags/en/Action/'
content = requests.get(url2).content
soup = BeautifulSoup(content,'lxml')

image_tags = soup.findAll('img')

for image_tag in image_tags:
    image_urls.append(image_tag.get('src'))
image_urls
```
