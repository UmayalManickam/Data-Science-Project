```python
doc = nlp("I prefer the morning flight through Denmark")                           
print('Input Text:\n',doc,'\n')                             
for chunk in doc.noun_chunks:
   print('"Text"=',chunk.text,'\t"Root text"=',chunk.root.text, '\t"Root dep"=',chunk.root.dep_,
          '\t"Root head text"=',chunk.root.head.text)    
```

```python
### TASK 6
```

```python
r=open('Raw_data_for_analysis.txt').read()
raw=re.sub('- ','',r);
Raw=nlp(raw)
print(Raw)
```

```python
#TOKENIZATION
print ([token.text for token in Raw]) 
```

```python
#Remove stop words from file
filtered_data2=[]                                                                                                                     
for word in Raw:                                                         
    if word.is_stop==False:                                               
        filtered_data2.append(word)                                        
print(filtered_data2)
```

```python
#lemmatization after removal of stop words      
for word in filtered_data2:                                                      
    print(word.text,'-> ',word.lemma_)  
```

```python
#POS tagging after removal of stop words.
for word in filtered_data2:
    print(word.text,'\t',word.pos_,'\t',word.tag_)
```

```python
#NER   
RR=nlp(str(filtered_data2))
displacy.render(RR, style = "ent",jupyter = True)                   
for ent in RR.ents:
    print(ent.text, '\tstart=',ent.start_char, '\tend=',ent.end_char, '\tLabel=',ent.label_) 
```
